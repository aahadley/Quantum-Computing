##Problem 1

Let $ B = \{| \psi_1 \rangle, ... , | \psi_d \rangle \}$ and $B' = \{ | \psi_1' \rangle , ... , | \psi_d' \rangle \}$ be two orthonormal bases of $\mathbb{C^d}$.  

The Orthonormal bases $B$ and $B'$ are called mutually unbiased if  

$$ 
| \langle \psi_i \rangle | \psi_j' \rangle |^2 = \frac{1}{d} 
$$

for all $i,j \epsilon [1, d]$.  

Show that the eigenbases of the pauli operators:  

$$ 
\sigma_x  = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}, \sigma_y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}, \sigma_z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} 
$$ are mutually unbiased.  

---
---
---


### Solution  

First, the eigenvalues $\lambda_1, \lambda_2$ of $\sigma_x$ can be calculated by solving the characteristic equation,  

$$
det(\sigma_x - \lambda I) = 
\begin{vmatrix} 
    \begin{bmatrix} 
        -\lambda & 1 \\ 
        1 & -\lambda 
    \end{bmatrix} 
\end{vmatrix} = \lambda^2 - 1 = \vec{0}.
$$ Thus,  

$$ 
\lambda_1 = 1 \\ 
\lambda_2 = -1
$$  

$$\\ %Not sure if there's a better way to add more whitepace$$

To obtain the eigenvector corresponding to $\lambda = 1$, let  

$$ 
\begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix} 
 \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} 
 = \vec{0}. 
 \implies \vec{v_{ \lambda = 1}} 
 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} .
$$  

Normalizing $\vec{v_{\lambda = 1}}$ yields 

$$
\vec{v_{\lambda=1}} = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
$$  

$$\\$$

Now for $\lambda = -1$, let

$$ 
\begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} 
 \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} 
 = \vec{0}. 
 \implies \vec{v_{ \lambda = 1}} 
 = \begin{bmatrix} 1 \\ -1 \end{bmatrix} .
$$

Normalizing $\vec{v_{\lambda = -1}}$ yields  
$$
\vec{v_{\lambda = -1}} = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)
$$  
$$\\$$

This leads to the conclusion that 
$$
B_x = \{ \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle),  \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)  \} 
$$
is an eigenbasis for $\sigma_x$.  

Proceeding similarly yields  

$$
B_y = \{ \frac{1}{\sqrt{2}}( |0 \rangle + i|1\rangle ), \frac{1}{\sqrt{2}}( |0 \rangle - i|1\rangle )  \}
$$
$$\\$$
$$
B_z  = \{ \frac{1}{\sqrt{2}}|0\rangle, \frac{1}{\sqrt{2}}|1\rangle \}
$$

---
---
Now we show that $B_x$, $B_y$, $B_z$ are mutually unbiased.  

As a preliminary, note that 
$$ 
\forall a, b \epsilon \mathbb{C}, |\langle a|b\rangle | = |\langle b | a \rangle| = |\langle a | b \rangle ^*| = |\langle b | a \rangle^*|. 
$$

Beginning first with $B_x$ and $B_z$,  

$$
\left| \langle \frac{1}{\sqrt{2} } (|0\rangle + |1\rangle ) | 0\rangle \right|^2 \\
= \left| \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \end{bmatrix}
\begin{bmatrix} 1 \\ 0 \end{bmatrix} \right|^2 \\
=\left| \frac{1}{\sqrt2} \right|^2 = \frac{1}{2}.
$$

and

$$
\left| \langle \frac{1}{\sqrt{2} } (|0\rangle - |1\rangle ) | 1\rangle \right|^2 \\
= \left|\frac{1}{\sqrt{2}} \begin{bmatrix} 1 & -1  \end{bmatrix} 
\begin{bmatrix} 0 \\ 1 \end{bmatrix} \right|^2 \\
= \left|\frac{-1}{\sqrt2}\right|^2 = \frac{1}{2}.
$$

It is clear from the preliminary result that any other applicable product of vectors in $B_x$ and $B_z$ will result in an absolute value of $\frac{1}{2}$. This result will also apply to the following computations.  

---

Continuing with $B_x$ and $B_y$,  

$$
\left| \langle \frac{1}{\sqrt2} ( |0 \rangle + |1\rangle) | \frac{1}{\sqrt2}(|0\rangle + i|1\rangle) \rangle \right|^2 \\
= \left| \frac{1}{\sqrt2}\langle |0\rangle + |1\rangle  |0\rangle + i|1\rangle \rangle \right|^2\\
=\left| \frac{1}{\sqrt2}\langle 1 | +i|1\rangle \rangle \right|^2 \\
= \left|\frac{i}{\sqrt2} \right|^2 = \frac{1}{2}.
$$  

and  

$$
\left| \langle \frac{1}{\sqrt2} ( |0 \rangle - |1\rangle) | \frac{1}{\sqrt2}(|0\rangle - i|1\rangle) \rangle \right|^2 \\
= \left| \frac{1}{\sqrt2}\langle |0\rangle - |1\rangle  |0\rangle - i|1\rangle \rangle \right|^2\\
=\left| \frac{1}{\sqrt2}\langle 1 | -i|1\rangle \rangle \right|^2 \\
= \left|\frac{i}{\sqrt2} \right|^2 = \frac{1}{2}.
$$

---
Now for $B_y$ and $B_z$,  

$$
\left| \langle \frac{1}{\sqrt{2} } (|0\rangle + i|1\rangle ) | 0\rangle \right|^2 \\
= \left| \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & i \end{bmatrix}
\begin{bmatrix} 1 \\ 0 \end{bmatrix} \right|^2 \\
=\left| \frac{1}{\sqrt2} \right|^2 = \frac{1}{2}.
$$  

and  

$$
\left| \langle \frac{1}{\sqrt{2} } (|0\rangle - i|1\rangle ) | 1\rangle \right|^2 \\
= \left| \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & -i \end{bmatrix}
\begin{bmatrix} 0 \\ 1 \end{bmatrix} \right|^2 \\
=\left| \frac{-i}{\sqrt2} \right|^2 = \frac{1}{2}.
$$  

Hence, $B_x$, $B_y$, and $B_z$ form mutually unbiased bases.

```
import numpy as np
from numpy import linalg as la

def get_eigenbasis(m):
    '''returns normalized eigenbasis of a matrix as a set of vectors'''
    
    e_val, e_vec = la.eig(m)
    return e_vec

def is_unbiased(v1, v2):
    '''returns True if v1 and v2 are unbiased'''
    
    if np.isclose( np.abs(np.vdot(v1, v2))**2 , 1/v1.shape[0]):
        return True
    
    return False

def is_mutually_unbiased(b1, b2, b3):
    '''Check if a set of three bases is mutually unbiased'''
    for i in b1:
        for j in b2:
            if not is_unbiased(i, j):
                return False
            
    for i in b1:
        for j in b3:
            if not is_unbiased(i, j):
                return False

    for i in b2:
        for j in b3:
            if not is_unbiased(i, j):
                return False
    
    return True


px = np.array([[0, 1],
               [1, 0]])

py = np.array([[0, -1j],
               [1j,  0]])

pz = np.array([[1,  0],
               [0, -1]])

is_mutually_unbiased(get_eigenbasis(px), get_eigenbasis(py), get_eigenbasis(pz))
    
    
```

## Problem 2

For $A, B \epsilon \mathbb{C}^{dxd}$, define 

$$
\langle A| B \rangle_{Tr} = Tr(A^{\dagger}B).
$$

Prove that the above map defines an inner product on the vector space $\mathbb{C}^{dxd}$.  

---
---
---
### Solution

#### Linearity in the second argument

$$
Tr \left(A^{\dagger} \sum_i \lambda_i B_i \right) = \sum_{i, j} \left( A^{*}_{i,j} \sum_i \lambda_iB_{i,j} \right) \\
= a^*_{1,1} \sum_i\lambda_i b_{1,1} + ... + a^*_{i,j}\sum_i\lambda_i b_{i,j} \\
= \sum_i \lambda_i (a^*_{1,1}b_{1,1} + ... + a^*_{i,j}b_{i,j})  \\
= \left(\sum_i \lambda_i \right)Tr(A^\dagger B).
$$

#### Conjugate Commutativity

$$
Tr(B^{\dagger} A) = Tr((B^\dagger A)^\dagger ) = Tr(A^\dagger B)).
$$

#### Non Negativity  

$$
Tr(A^\dagger A) = a^*_{1,1}a_{1,1} + ... +  a^*_{i,j}a_{i,j} .
$$

For any $z \epsilon \mathbb{C}, $ where $  \alpha , \beta \epsilon \mathbb{R} $ ,

$$
zz^* = (\alpha + \beta i)(\alpha - \beta i) = \alpha ^2 + \beta ^2
$$  

So,  

$$
Tr(A^\dagger A) \geq 0
$$ with equality if and only if $A = 0$.

## Problem 3

Let $X, Z \epsilon \mathbb{C}^{dxd}$ be  

$$
X = \sum_{k=0}^{d-1} = |k + 1\rangle \langle k |  \\
Y = \sum_{k=0}^{d-1} = \omega ^l |l \rangle \langle l |
$$

Where the addition is modulo $k+1$ and $\omega = e^{\frac{2 \pi i}{d}}$ is a primitive $d^{th}$ root of unity.  

Show that the $d^2$ matrices $M^{(a, b)} = X^a Z^b$  

where $a, b \epsilon \{0, ..., d-1\}$ form an orthonormal basis with respect to the trace inner product for all pairs.  

--

```
import numpy as np
from numpy import linalg as la

def get_vec(n, dim):
    print(n)
    l = [0 for i in range(dim)]
    l[n] = 1
    
    return np.array(l)

def get_X(d):
    
    sum = np.array([0 for i in range(d**2)])
    for k in range(d):
        sum += np.kron(get_vec((k+1), d), get_vec(k, d))
    
    return sum

def get_Z(d):
    
    w = np.exp((2 * np.pi * 1j)/d)
    sum = np.array([0 for i in range(d**2)])
    for l in range(d):
        sum = np.add(sum, np.kron(w**l*get_vec(l, d), get_vec(l, d)), casting='unsafe')
    
    return sum

def get_error_basis(d):
    
    l = []
    
    for a in range(d):
        for b in range(d):
            l.append(np.dot(la.matrix_power(a, get_X(d)), la.matrix_power(b, get_Z(d))))
    
    return l

def trace_inner_product(a, b):
    return np.trace(np.vdot(a.conj().T, b))
        

l = get_error_basis(int(input()))
l_tip = []

for i in l:
    for j in l:
        l_tip.append(trace_inner_product(i, j))
        
```
